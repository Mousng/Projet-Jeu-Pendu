```python
import tkinter as tk
import random
from tkinter import messagebox
```

```python

#création de l'interface

def jouer_au_pendu() :
    global fenetre, nb_tentatives, demander_rejouer, end, score
    nb_tentatives = 7
    end = False
    #score = 0

#création de la fenetre

    fenetre = tk.Tk()
    fenetre.title("Jeu du Pendu")

# Mots à deviner

    mots = ['PARIS', 'MARSEILLE', 'BORDEAUX', 'INFORMATIQUE','LILLE','VERSAILLES']

#fonction qui permet de choisir un mot aléatoirement

    def choisir_un_mot():
        return random.choice(mots)

#initialisation de commande pour cacher le mot secret et la fonction recharger un mot

    mot_secret = choisir_un_mot()
    mot_a_trouve = ['*'] * len(mot_secret)
    print(mot_secret)
    label_mot = tk.Label(fenetre, text=' '.join(mot_a_trouve))
    label_mot.pack()

    lettre_alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

    boutons_lettres = []
    for j, lettre in enumerate(lettre_alphabet):
        bouton = tk.Button(fenetre, text=lettre, command=lambda l=j: choisir_lettre(l))
        bouton.pack(side=tk.LEFT,ipadx= 2, ipady=2)
        boutons_lettres.append(bouton)

#TOUTES LES GRANDES FONCTION DE L'INTERFACE


    #fonction qui associe les boutons aux lettres_alphabet

    def choisir_lettre(j):
        global nb_tentatives, end,score
        if end :
            return 
        #print("Lettre cliquée:", lettre_alphabet[j])
        if lettre_alphabet[j] in mot_secret:
            for i in range(len(mot_secret)):
                if mot_secret[i] == lettre_alphabet[j]:
                    mot_a_trouve[i] = lettre_alphabet[j]
            label_mot.config(text=" ".join(mot_a_trouve))
            if "*" not in mot_a_trouve:
                victoire_defaite()  # Appel de la fonction victoire_defaite si le joueur a gagné
                continuer_quitter() # Appel de la fonction pour proposer au joueur de rejouer
                #afficher_score()
                #score += 1
                end = True
            
                
        else:
            nb_tentatives -= 1
            #score += 1
            dessiner_pendu(nb_tentatives)
            nb_tentatives_label.config(text="ESSAIS RESTANTS:" + str(nb_tentatives))
            if nb_tentatives == 0:
                victoire_defaite() #Appel de la fonction victoire_defaite une deuxieme fois si le joueur a perdu
                continuer_quitter() # Appel de la fonction pour proposer au joueur de rejouer
                #afficher_score()
                end= True
        boutons_lettres[j].config(state="disabled")


    #condition de victoire et de defaite 
    def victoire_defaite():
        if "*" not in mot_a_trouve :
            gagne_label = tk.Label(fenetre, text="BRAVO : GAGNÉ !")
            gagne_label.pack(ipadx= 2, ipady=2)
        else :
            perdu_label = tk.Label(fenetre, text="DOMMAGE : PERDU, LE MOT ÉTAIT " + mot_secret)
            perdu_label.pack(ipadx= 2, ipady=2)
    
    #fonction qui permet de quitter la fenetre
    def fermer_fenetre():
        fenetre.destroy()

    
    
    #fonction qui demande au joueur s'il veut continuer

    def non_rejouer() :
        fenetre.destroy()

    def continuer_quitter():
            choix = tk.Label(fenetre,text="VOULEZ-VOUS REJOUER ?")
            choix.pack()
            reponse_oui= tk.Button(fenetre, text='OUI', command=recharger_un_nv_mot)
            reponse_oui.pack()
            reponse_non = tk.Button(fenetre, text= "NON", command= non_rejouer)
            reponse_non.pack()
            
    
    #fonctionne qui dessine le pendu

    def dessiner_pendu(nb_tentatives):
        if nb_tentatives == 6:
             canvas.create_line((0, 400), (300, 400), fill="black", width=2) 
             canvas.create_line((55, 110), (55, 400), fill="black", width=2)
             canvas.create_line((55, 135), (77, 110), fill="black", width=2)
             canvas.create_line((55, 110), (175, 110), fill="black", width=2) 
        elif  nb_tentatives == 5:
               canvas.create_line((175, 110),(175, 150), fill="black", width=2)
               canvas.create_oval((150, 150), (200,200), fill="red", width=2)  
        elif nb_tentatives == 4:
            canvas.create_line((175, 180), (175, 300), fill="black", width=2)
        elif nb_tentatives == 3:
            canvas.create_line((175, 200), (150, 250), fill="black", width=2)
        elif nb_tentatives ==2:
            canvas.create_line((175, 200), (200, 250), fill="black", width=2)
        elif nb_tentatives == 1:
            canvas.create_line((175, 300), (150, 350), fill="black", width=2)
        else:
            nb_tentatives == 0
            canvas.create_line((175, 300), (200, 350), fill="black", width=2)
        indication_mot()

    #fonction qui propose une indication au joueur
    
    def indication_mot():
        indication = ' '
        if mot_secret =="PARIS" and (nb_tentatives == 3 or nb_tentatives== 2 or nb_tentatives== 1):
            indication = "INDICATION : C'EST UNE CAPITALE TRES CONNU"
        elif mot_secret == "BORDEAUX" and (nb_tentatives == 3 or nb_tentatives== 2 or nb_tentatives== 1) :
            indication = "INDICATION : C'EST UNE VILLE SUD-OUEST DE LA FRANCE "
        elif mot_secret == "INFORMATIQUE" and (nb_tentatives == 3 or nb_tentatives== 2 or nb_tentatives== 1) :
            indication ="INDICATION : PERMET D'APPRENDRE LA PROGRAMMATION"
        elif mot_secret == "LILLE" and (nb_tentatives == 3 or nb_tentatives== 2 or nb_tentatives== 1):
            indication ="INDICATION : C'EST UNE GRANDE DU NORD DE LA FRANCE"
        elif mot_secret == "MARSEILLE" and (nb_tentatives == 3 or nb_tentatives== 2 or nb_tentatives== 1) :
            indication ="INDICATION : VILLE PORTIERE DE LA FRANCE"
        elif mot_secret == "VERSAILLES" and (nb_tentatives == 3 or nb_tentatives== 2 or nb_tentatives== 1):
            indication = "INDICATION : VILLE DE LOUIS XIV"
        indication_label.config(text=indication)
        print(indication)
    

    #Fonction qui charge un nouveau mot
    
    def recharger_un_nv_mot():
        mot_secret = choisir_un_mot()
        mot_à_trouve = ['*'] * len(mot_secret)
        label_mot.config(text=' '.join(mot_à_trouve))
        label_mot.pack()
        print(mot_secret)
        jouer_au_pendu()
   
    # fonction qui affiche une fenêtre pop-up avec les règles du jeu

    def afficher_aide():
        messagebox.showinfo("Aide", "Règles du jeu :\n\n1. Cliquez sur une lettre pour faire une tentative.\n2. Si la lettre est dans le mot secret, vous gagnez immédiatement.\n3. Si la lettre n'est pas dans le mot secret, le nombre d'essais restants diminue.\n4. Vous perdez si le nombre d'essais autorisé est atteint.")
    

#TOUS LES BOUTONS ET LABEL DE L'INTERFACE
    indication_label = tk.Label(fenetre, text='')
    indication_label.pack(anchor="center")

    canvas = tk.Canvas(fenetre, bg="white", height=400, width=400)
    canvas.pack(anchor="ne") #création du canvas

    nb_tentatives_label = tk.Label(fenetre, text='ESSAIS RESTANTS: ' + str(nb_tentatives))
    nb_tentatives_label.pack(anchor="center") #le boutton des nombres tentatives restantes

    
    bouton_quitter = tk.Button(fenetre, text="QUITTER LE JEU", command=fermer_fenetre)
    bouton_quitter.pack(anchor="nw",ipadx= 2, ipady=2)#le boutton quitter

    
    bouton_aide = tk.Button(fenetre, text='AIDE AU JEU', command=afficher_aide)
    bouton_aide.pack(side="left",ipadx= 2, ipady=2) #le boutton aide

    bouton_nv= tk.Button(fenetre, text="CHARGER UN NOUVEAU MOT", command=recharger_un_nv_mot)
    bouton_nv.pack(side="left",ipadx= 2, ipady=2) #boutton pour le nouveau mot


            




       
    












    
    fenetre.mainloop()
jouer_au_pendu()
```
